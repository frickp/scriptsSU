```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
```

```python
qc = pd.read_csv('151020_PC9_3dErl.csv')
```

```python
qc
```

```python
qc2 = pd.melt(qc,['Condition','Cell_line','Sample'])
qc2['id']=qc2['Condition'] + '_' + qc2['Cell_line']
qc2
```

```python
fig=plt.figure(figsize=(6,4),dpi=1600)
ax=plt.subplot(111)
#ax.set_xlabel('SampleID',fontsize=14)
sns.barplot(x='id',y='value',hue='variable',data=qc2)
sns.axlabel('SampleID','Mean value',fontsize=14)
plt.show()
fig.savefig('151020_PC9_3dErl',dpi=1600,bbox_inches='tight')
```
